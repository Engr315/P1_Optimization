## Testing the Star Plotting Function

## Testing Plotting Many stars
So far this is too fast. Since the number of points computed is so low. I will increase this soon, just wanted to get something up. Or perhaps we will simply utilize more stars. It plots 1000 stars in about 0.01ss on my machine

```python
num_rand_stars = 1000

inputs = [(rnd.randint(0,10000), rnd.randint(0,10000), rnd.randint(0, 300)) for _ in range(num_rand_stars)]

start = time.perf_counter()
list_of_stars = bad_star_instance.draw_many_star(inputs)
stop = time.perf_counter()

plt.figure(figsize=(5,5))
for star in list_of_stars:
    x_s, y_x = zip(*star)
    plt.plot(x_s, y_x)
plt.xlim([0,10000])
plt.ylim([0,10000])
plt.show()
print(f"Total Time to compute {num_rand_stars} stars: {stop-start}")
```
